```python
# just to view if we have data already?
df.head()
```

```python
print(df.embark_town.unique())  #'printed'to view values in embk_tn column
print(df[(df.embark_town=="Southampton")]) #'printd'to view only southamp in embk_tn
df[(df.embark_town=="Southampton")].shape # to view shape of the data
```

### *Note:- Below "1, 2, 3, 4" are different possibilities. Best is to see and analyze according to your data and what is fit to you. Overall, it is easy. Finally" some thing single vs single ya  singe but outside the block of multiple having (&) comes alone".

```python
# 1
# here we used 'or', it means sex column se only female or embk_town se only southampton
# hon End result is mixed gender and mixed area wise.
# 'or,|' takes data separate separate but at the end it becomes mixed.
df[(df.embark_town=="Southampton") |
   (df.sex=="female")]
```

```python
#1 
# here we used 'and', it means sex se only female and embk_town se only southampton
# hon. End result is gender only female, area only southamptn bcz of 'and'.
# 'and, &' takes data separate but at the end keeps also separate.
df[(df.embark_town=="Southampton") &
   (df.sex=="female")]
```

```python
# 2
df[((df.embark_town=="Southampton") |
   (df.embark_town=="Queenstown"))|
   (df.sex=="female")]
```

```python
# 2
# here we choose 2 areas by "or", but gender only female by "and".
# so "female" is with (&) single but out of block, so it comes alone not mixed.
df[((df.embark_town=="Southampton") |
   (df.embark_town=="Queenstown"))&
   (df.sex=="female")]
```

```python
# 3
# Here also southhampton comes alone like above.
df[(df.embark_town=="Southampton") &
   ((df.sex=="female")|(df.sex=="male"))]
```

```python
# 3
df[(df.embark_town=="Southampton") |
   ((df.sex=="female")|(df.sex=="male"))]
```

```python
# 4
# here we choose 2 areas by "or", gender also two male-female by "or". But in btween
# we used (&) End result is with two areas bcz of (&).
df[((df.embark_town=="Southampton") |
   (df.embark_town=="Queenstown"))&
   ((df.sex=="female")|(df.sex=="male"))]
```

```python
# 4
# here we choose 2 areas by "or", gender also two male-female by "or". But in btween
# we used also (|) End result is with three areas bcz of (|).
df[((df.embark_town=="Southampton") |
   (df.embark_town=="Queenstown"))|
   ((df.sex=="female")|(df.sex=="male"))]
```

```python
# to view only Queenstown in embk_tn,mostly like this formula/ df[(df.embark_town=="Southampton")]
df[df.embark_town.isin(["Queenstown"])].head()
```

```python
# same like uper but here you can add 2 or even more
df[df.embark_town.isin(["Queenstown", "Southampton"])].head()
```

# 16- Splitting a string into multiple columns

```python
# import library
import pandas as pd

df = pd.DataFrame({"name": ["Hadia Fatima", "Haseeb Ahmed", "Zain Abbas", "Abeer Abbas", "Manahil Fatima"],
              "location": ["Jaranwala, Pakistan", "Jhumra, Pakistan", "Aachen, Germany", "Jaranwala, Pakistan", "Jhumra, Pakistan"]})
df
```

```python
#* splitting 'name column' into two columns
#* could be split and convert into new columns in a single formula as below
df.name.str.split(" ", expand=True)
```

```python
# adding those splits into new column
# expand wali braket me iss"" qomay me space ya qoma wo hi dena hy jo actual halat me use huwa, qoma ya space e.t.c.
df[["first_name", "last_name"]] = df.name.str.split(" ", expand=True)
df
```

```python
# splitting & adding 'location' column into 2 columns in a single formula
df[["city", "country"]] = df.location.str.split(", ", expand=True)
df
```

```python
# finally refining the columns which only we need/Refine data manipulation
df = df[["first_name", "last_name", "city", "country"]]
df
```
